## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

### Loading the Json data

```python
with open('data/world_bank_projects.json', 'r') as file:
    json_data = json.load(file)
json_data[0]
```

### Normaling the Json

```python
world_bank_df = json_normalize(json_data, 'mjtheme_namecode', [['_id','$oid'], 'countrycode', 'countryname', 'project_name'])
world_bank_df.columns = ['mjtheme_code', 'mjtheme_name', 'id', 'country_code','country_name', 'project_name']
world_bank_df.head()
```

### Processing dataframe to fill missing names.

```python
clean_world_bank_df = world_bank_df

# Converting data type of mjtheme_code to int
clean_world_bank_df['mjtheme_code'] = pd.to_numeric(clean_world_bank_df['mjtheme_code'])

# Sorting world_bank_df with respect to mjtheme_code column
clean_world_bank_df = clean_world_bank_df.sort_values(['mjtheme_code','mjtheme_name'])
clean_world_bank_df = clean_world_bank_df.reset_index(drop=True)

# Fill missing/empty mjtheme_name with backward fill
clean_world_bank_df = clean_world_bank_df.replace('', np.nan)
clean_world_bank_df['mjtheme_name'] = clean_world_bank_df['mjtheme_name'].fillna(method='bfill')

clean_world_bank_df.head()
```

#### Find the 10 countries with most projects

```python
# Considering subset of dataframe for answering above question
project_df = clean_world_bank_df[['id', 'country_name', 'project_name']]
project_df = project_df.drop_duplicates()
project_df = project_df.set_index('id')

# Grouping data by country_code
country_df = project_df.groupby('country_name').count()
country_df.columns = ['total_projects']
country_df = country_df.sort_values('total_projects', ascending=False)

# 10 Countries with most projects are below
top_10_country_df = country_df.head(10)
top_10_country_df
```

```python
# Considering only top 20 Countries to display
top_country_df = country_df.head(20)

# Visualizing the 'country' with respect to 'total projects'
ax = top_country_df.plot(kind='bar', ylim=[0,20], title='Countries with total projects')
ax.set_ylabel('Total Projects')
ax.set_xlabel('Country')
plt.show()
```

#### Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
# Considering subset of dataframe for answering above question
mjtheme_df = clean_world_bank_df[['mjtheme_name', 'project_name']]
mjtheme_df = mjtheme_df.drop_duplicates()

# Grouping data by mjtheme_name
mjtheme_df = mjtheme_df.groupby('mjtheme_name').count()
mjtheme_df.columns = ['total_projects']
mjtheme_df = mjtheme_df.sort_values('total_projects', ascending=False)

# 10 Countries with most projects are below
top_10_mjtheme_df = mjtheme_df.head(10)
top_10_mjtheme_df
```

```python
# Visualizing the 'country' with respect to 'total projects'
ax = mjtheme_df.plot(kind='bar', ylim=[0,160], title='Major Project Theme')
ax.set_ylabel('Total Projects')
ax.set_xlabel('Major Theme')
plt.show()
```
